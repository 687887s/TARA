```python
import pandas as pd
import numpy as np
```

```python
filepath = "../data/merged_training_ps1.csv"
compfilepath = "../data/training/training_data.csv"
```

```python
df = pd.read_csv(filepath)
comp_df = pd.read_csv(compfilepath)
comp_df = comp_df[['ra', 'dec', 'spec_redshift', 'spec_redshift_error']]

comp_df = comp_df[np.isin(comp_df['ra'].values , df['orig_ra'].values)]
comp_df = comp_df.sort_values(by='ra')
df = df.sort_values(by='orig_ra')
df = df[['raMean', 'decMean', 'gMeanKronMag', 'rMeanKronMag', 'iMeanKronMag', 'zMeanKronMag', 'yMeanKronMag', 'gMeanKronMagErr', 'rMeanKronMagErr', 'iMeanKronMagErr', 'zMeanKronMagErr', 'yMeanKronMagErr']]
df['spec_redshift'] = comp_df['spec_redshift']
df['spec_redshift_error'] = comp_df['spec_redshift_error']
```

```python
df.to_csv(filepath)
```

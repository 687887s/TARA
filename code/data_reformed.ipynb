```python
import pandas as pd
import numpy as np
```

```python
filepath = "../data/merged_training_ps1.csv"
compfilepath = "../data/training/training_data.csv"
```

```python
comp_df = pd.read_csv(compfilepath)
comp_df
```

```python
df = pd.read_csv(filepath)
comp_df = pd.read_csv(compfilepath)
comp_df = comp_df[['ra', 'dec', 'spec_redshift', 'spec_redshift_error', 'zWarning']]

# df = df[['objID','raMean', 'decMean', 'gMeanKronMag', 'rMeanKronMag', 'iMeanKronMag', 'zMeanKronMag', 'yMeanKronMag', 'gMeanKronMagErr', 'rMeanKronMagErr', 'iMeanKronMagErr', 'zMeanKronMagErr', 'yMeanKronMagErr', 'orig_ra', 'orig_dec']]
df = pd.merge(df, comp_df, left_on=['orig_ra', 'orig_dec'], right_on=['ra', 'dec'], how='left')
df.drop(columns=['ra', 'dec'], inplace=True)
```

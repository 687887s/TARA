```python
import pandas as pd
import glob
```

```python
datas = glob.glob('../data/catalog/ps1/*.csv')
z_phot = pd.read_csv('ps1_estimate_xgb.csv')
z_phot_filtered = pd.read_csv('ps1_estimate_filtered_xgb.csv')
```

```python
for target in datas:
    df = pd.read_csv(target)
    df = pd.merge(df, z_phot, left_on='objID', right_on='objID', how='left')
    df[pd.isna(df['z_phot']) == False].to_csv('../data/catalog/ps1/xgb/result' +target.split('../data/catalog/ps1')[1])
```

```python
for target in datas:
    df = pd.read_csv(target)
    df = pd.merge(df, z_phot_filtered, left_on='objID', right_on='objID', how='left')
    df[pd.isna(df['z_phot']) == False].to_csv('../data/catalog/ps1/xgb/result_filtered' +target.split('../data/catalog/ps1')[1])
```

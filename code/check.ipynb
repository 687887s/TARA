```python
import glob
import pandas as pd
import numpy as np
from astropy.cosmology import Planck18
```

```python
ls = np.array(glob.glob('../data/redshift/*.csv'))
comp = pd.read_csv('../all_sample_20260702.csv')
comp_ra = comp['paliya_RA'].values
comp_dec = comp['paliya_DEC'].values
comp_z = comp['paliya_Z'].values
```

```python
z_value = comp['paliya_Z'].values
comp = comp.rename(columns = {"paliya_RA":"RA", "paliya_DEC":"DEC"})

distance = Planck18.comoving_distance(z_value).value
size_ang = np.rad2deg(np.arcsin(0.5/2/distance))
size = np.astype(np.rad2deg(np.arcsin(0.5/2/distance)*2)*3600*4, 'int')
```

```python
compare_name = np.array(['../data/redshift\\{:4f}-{:4f}-{:4f}.csv'.format(ra, dec, z) for ra, dec, z in zip(comp_ra, comp_dec, size_ang*60)])
```
